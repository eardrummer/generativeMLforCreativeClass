### 3) Train your own model (simplest way)
I use the seinfeld corpus which was created by scraping the scripts of all the episodes.
https://www.kaggle.com/luongleanstocode/seinfeld-text-corpus/download

```python
textgen4 = textgenrnn()
```

```python
textgen4.train_from_file('./seinfeld-text-corpus/corpus.txt', num_epochs=1)
```
